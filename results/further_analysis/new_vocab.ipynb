The purpose of this analysis is to see whether our question generation model generates any new vocabulary words - words not seen during training. This is plausible due to GPT-2's tokenization, but the strong effect of finetuning may constrain our models to observed vocabulary words. **We only analyze the generated outputs which we calculated statistics over for the paper**

```python
with open("../generation_results/generation_outputs/french_penalty","rb") as f:
    french_data = f.readlines()
with open("../generation_results/generation_outputs/spanish_penalty","rb") as f:
    spanish_data = f.readlines()
with open("../../data/raw_data/fr_en.slam.20190204.train","rb") as f:
    french_train = f.readlines()
with open("../../data/raw_data/fr_en.slam.20190204.dev","rb") as f:
    french_dev = f.readlines()
with open("../../data/raw_data/fr_en.slam.20190204.test","rb") as f:
    french_test = f.readlines()
with open("../../data/raw_data/es_en.slam.20190204.train","rb") as f:
    spanish_train = f.readlines()
with open("../../data/raw_data/es_en.slam.20190204.dev","rb") as f:
    spanish_dev = f.readlines()
with open("../../data/raw_data/es_en.slam.20190204.test","rb") as f:
    spanish_test = f.readlines()
```

Concatenate all Duolingo datasets to create the entire vocab. Takes a while to run :(

```python
french_vocab = " ".join([str(s) for s in french_train+french_dev+french_test]).lower()
spanish_vocab = " ".join([str(s) for s in spanish_train+spanish_dev+spanish_test]).lower()
```

```python
french_words = str(french_data[-10]).split(" ")
spanish_words = str(spanish_data[-10]).split(" ")

def check_word(words):
    for w in words:
        #this checks if the word doesn't appear at all in the Duolingo released datasets
        if(w.isalnum() and w.lower() not in spanish_vocab and w.lower() not in french_vocab):
            print(w.lower())

print("Spanish Question Generation Model New Words:")
check_word(spanish_words)
print("\n French Question Generation Model New Words:")
check_word(french_words)
```

As can be observed, many of the "new words" are actually conjunctions that are not real English words, or phrases such as "timebordeaux". However, we do observe new vocabulary words such as **overlap**, **biology**, **endless**, **adheres**, **endangering**, **damaged**. We search the dataset to see if there are words with shared roots as vocabulary in the Duolingo dataset, that may have been generated due to the tokenization / stemming process in language models.
